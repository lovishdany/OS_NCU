```python
import pandas as pd
import math
```

```python
print("Shortest Job First Scheduling")
n=int(input("Enter number of processes: "))
burst_time=[]
arrival_time=[]
tat=[0]*n
completion_time=[0]*n
time=0
waiting_time=[0]*n



for i in range(0,n):
  a=int(input(f"Enter Burst Time for process {i+1}:"))
  burst_time.append(a)
  arrival_time.append(i)

temp=burst_time.copy()
time=time + burst_time[0]
completion_time[0]=time
t=completion_time[0]-arrival_time[0]
tat[0]=t
w=tat[0]-burst_time[0]
waiting_time[0]=w
temp[0]=math.inf
for i in range(1,n):
  m=min(temp[1:time])
  index=burst_time.index(m)
  time=time + burst_time[index]
  completion_time[index]=time
  t=completion_time[index]-arrival_time[index]
  tat[index]=t
  w=tat[index]-burst_time[index]
  waiting_time[index]=w
  temp[index]=math.inf

df=pd.DataFrame(list(zip(arrival_time, burst_time,completion_time, tat, waiting_time)),
               columns =['Arrival time', 'Burst time','Completion' , 'TAT', 'Waiting time'])

df
```

```python
print("Shortest Remaining Time First Scheduling")
n=int(input("Enter number of processes: "))
burst_time=[]
arrival_time=[]
tat=[0]*n
completion_time=[0]*n
time=0
waiting_time=[0]*n



for i in range(0,n):
  a=int(input(f"Enter Burst Time for process {i+1}:"))
  burst_time.append(a)
  arrival_time.append(i)


remaining_time=burst_time.copy()
iteration=sum(burst_time)
print(remaining_time)

time=time + 1
remaining_time[0]=remaining_time[0]-1


for i in range(0,iteration+n):
  m=min(remaining_time[0:time+1])
  index=remaining_time.index(m)
  if(remaining_time[index]==0):
    remaining_time[index]=math.inf
    completion_time[index]=time
    t=completion_time[index]-arrival_time[index]
    tat[index]=t
    w=tat[index]-burst_time[index]
    waiting_time[index]=w
  else:
    time=time+1
    remaining_time[index]=remaining_time[index]-1

df=pd.DataFrame(list(zip(arrival_time, burst_time,completion_time, tat, waiting_time)),
               columns =['Arrival time', 'Burst time','Completion' , 'TAT', 'Waiting time'])

df
```
