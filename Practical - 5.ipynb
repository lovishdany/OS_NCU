```python
print("FIRST COME FIRST SERVE SCHEDULING")
print("")
n= int(input("Enter number of processes : "))

at = []
bt = []
processes = []
for i in range(n):
  a = int(input("Enter arrival time of process P"+str(i)+": "))
  b = int(input("Enter burst time of process P"+str(i)+": "))

  at.append(a)
  bt.append(b)
  processes.append("P" + str(i))
print("")
print(at)
print(bt)
print(processes)
```

```python
import pandas as pd

data = {
    "Process" : processes,
    "Arrival Time" : at,
    "Burst Time" : bt
}

df = pd.DataFrame(data)
df
```

```python
df1 = df.sort_values(by=['Arrival Time']).reset_index(drop = True)
df1
```

```python
df1['Wait Time'] = ''
df1['TurnAround Time'] = ''
```

```python
for i in df1.index:
  if i == 0:
    df1['Wait Time'][i] = 0
  else:
    df1['Wait Time'][i] = df1['Burst Time'][i-1] + df1['Wait Time'][i-1]

df1
```

```python
for i in df1.index:
  df1['TurnAround Time'] = df1['Wait Time'] + df1['Burst Time']

df1
```

```python
Total_WT = sum(df1['Wait Time'])
print("Total Waiting Time = ",Total_WT)
print("Average Waiting Time = ",Total_WT / n)
```

```python
Total_TAT = sum(df1['TurnAround Time'])
print("Total TurnAround Time = ",Total_TAT)
print("Average TurnAround Time = ",Total_TAT / n)
```
